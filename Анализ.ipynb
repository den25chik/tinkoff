```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### Импортируем таблицу

```python
df = pd.read_excel('dz-z2.xls')
df
```

#### use_dt для анализа не очень подходит, добавим столбец флагов использования сертификата

```python
df['use'] = ((pd.isnull(df.use_dt))==False)*1
df
```

#### Посмотрим на конверсии по различным сертификатам:

```python
df1 = df.groupby('card_nm')['use'].sum().to_frame()
df2 = df.groupby('card_nm')['use'].count().to_frame()
df_conv = pd.merge(df2, df1, left_on='card_nm', right_on='card_nm')
df_conv.rename(columns={'use_x': 'issued', 'use_y': 'used'}, inplace=True)
df_conv['issued_to_used'] = (df_conv['used']/df_conv['issued'])*100
df_conv = df_conv.round({'issued_to_used': 0})
df_conv
```

#### Видно, что лучше всего используются сертификаты Gamma -- 36% использования. Меньше всего используют сертификаты Beta -- 7% использования. Посмотрим использование сертификатов в динамике.

#### Создадим отдельные подтаблицы для каждого месяца. Для апреля не создаем, т.к. он еще не закончился.

```python
df_nov = df[(df.receive_dt >= '2018-11-01') & (df.receive_dt < '2018-12-01')]
df_dec = df[(df.receive_dt >= '2018-12-01') & (df.receive_dt < '2019-01-01')]
df_jan = df[(df.receive_dt >= '2019-01-01') & (df.receive_dt < '2019-02-01')]
df_feb = df[(df.receive_dt >= '2019-02-01') & (df.receive_dt < '2019-03-01')]
df_mar = df[(df.receive_dt >= '2019-03-01') & (df.receive_dt < '2019-04-01')]
```

#### Ноябрь

```python
df1 = df_nov.groupby('card_nm')['use'].sum().to_frame()
df2 = df_nov.groupby('card_nm')['use'].count().to_frame()
df_nov_conv = pd.merge(df2, df1, left_on='card_nm', right_on='card_nm')
df_nov_conv.rename(columns={'use_x': 'issued', 'use_y': 'used'}, inplace=True)
df_nov_conv['issued_to_used'] = (df_nov_conv['used']/df_nov_conv['issued'])*100
df_nov_conv = df_nov_conv.round({'issued_to_used': 0})
df_nov_conv
```

#### Декабрь

```python
df1 = df_dec.groupby('card_nm')['use'].sum().to_frame()
df2 = df_dec.groupby('card_nm')['use'].count().to_frame()
df_dec_conv = pd.merge(df2, df1, left_on='card_nm', right_on='card_nm')
df_dec_conv.rename(columns={'use_x': 'issued', 'use_y': 'used'}, inplace=True)
df_dec_conv['issued_to_used'] = (df_dec_conv['used']/df_dec_conv['issued'])*100
df_dec_conv = df_dec_conv.round({'issued_to_used': 0})
df_dec_conv
```

#### Январь

```python
df1 = df_jan.groupby('card_nm')['use'].sum().to_frame()
df2 = df_jan.groupby('card_nm')['use'].count().to_frame()
df_jan_conv = pd.merge(df2, df1, left_on='card_nm', right_on='card_nm')
df_jan_conv.rename(columns={'use_x': 'issued', 'use_y': 'used'}, inplace=True)
df_jan_conv['issued_to_used'] = (df_jan_conv['used']/df_jan_conv['issued'])*100
df_jan_conv = df_jan_conv.round({'issued_to_used': 0})
df_jan_conv
```

#### Февраль

```python
df1 = df_feb.groupby('card_nm')['use'].sum().to_frame()
df2 = df_feb.groupby('card_nm')['use'].count().to_frame()
df_feb_conv = pd.merge(df2, df1, left_on='card_nm', right_on='card_nm')
df_feb_conv.rename(columns={'use_x': 'issued', 'use_y': 'used'}, inplace=True)
df_feb_conv['issued_to_used'] = (df_feb_conv['used']/df_feb_conv['issued'])*100
df_feb_conv = df_feb_conv.round({'issued_to_used': 0})
df_feb_conv
```

#### Март

```python
df1 = df_mar.groupby('card_nm')['use'].sum().to_frame()
df2 = df_mar.groupby('card_nm')['use'].count().to_frame()
df_mar_conv = pd.merge(df2, df1, left_on='card_nm', right_on='card_nm')
df_mar_conv.rename(columns={'use_x': 'issued', 'use_y': 'used'}, inplace=True)
df_mar_conv['issued_to_used'] = (df_mar_conv['used']/df_mar_conv['issued'])*100
df_mar_conv = df_mar_conv.round({'issued_to_used': 0})
df_mar_conv
```
