```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv', header=0, sep=';')
df
```

```python
#ÐœÐµÐ½ÑÐµÐ¼ Ñ„Ð¾Ñ€Ð¼Ð°Ñ‚ Ð´Ð°Ñ‚Ñ‹
df['game_dttm'] = pd.to_datetime(df['game_dttm'])
df['game_dttm'] = df['game_dttm'].dt.strftime('%Y/%m/%d')
df
```

```python
#Ð¡Ð¾Ð·Ð´Ð°Ð½Ð¸Ðµ Ñ‚Ð°Ð±Ð»Ð¸Ñ†Ñ‹ Ð´Ð»Ñ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ¾Ð²
df1 = df[['game_dttm', 'game_flg', 'finish_flg']]
df2 = df1.groupby('game_dttm')[['game_flg', 'finish_flg']].sum()
#ÐŸÐ¾Ð´ÑÑ‡ÐµÑ‚ ÐºÐ¾Ð½Ð²ÐµÑ€ÑÐ¸Ð¸
df2['game_to_finish'] = (df2['finish_flg']/df2['game_flg'])*100
#
df2.rename(columns={'game_flg': 'games', 'finish_flg': 'finished_games'}, inplace=True)
df2
```

```python
#ÐŸÐ¾ÑÑ‚Ñ€Ð¾ÐµÐ½Ð¸Ðµ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ¾Ð²

#Ð”Ð»Ñ Ð»ÑƒÑ‡ÑˆÐµÐ¹ Ð²Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ð¸
game_min = df2['games'].min()
game_max = df2['games'].max()
#

fig, ax = plt.subplots(figsize=(16, 8))
ax2 = ax.twinx()
blue = plt.cm.Blues(0.5)
red = plt.cm.Reds(0.5)
df2['games'].plot(kind='bar', color = blue, ax = ax, ylim = (game_min-50,game_max+50))
df2['game_to_finish'].plot(color = red, lw = 2, ax = ax2)

ax.set_ylabel('ÐšÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð¸Ð³Ñ€')
ax2.set_ylabel('ÐŸÑ€Ð¾Ñ†ÐµÐ½Ñ‚ Ð·Ð°Ð²ÐµÑ€ÑˆÐµÐ½Ð½Ñ‹Ñ… Ð¸Ð³Ñ€')
h1, l1 = ax.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax.legend(h1+h2, l1+l2)
```
